```python
def check_primality(number):
    for i in range(2, number):
        if number%i == 0:
            return False
    return True
```

```python
primes = []
for i in range(100):
    if check_primality(i):
        primes.append(i)
```

```python
def generate_DH_values(generator, prime, parameter_a, parameter_b):
    
    val_a = (generator ** parameter_a) % prime
    val_b = (generator ** parameter_b) % prime
    
    
    print("{}, A sends the parameter {}".format(parameter_a, val_a))
    print("{}, B sends the parameter {}".format(parameter_b, val_b))
    
    print("calculated shared secret key is {}".format((val_a**parameter_b)%prime))
    print("calculated shared secret key is {}".format((val_b**parameter_a)%prime))
    
    print("-"*30)
```
